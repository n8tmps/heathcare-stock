```python
#!pip install pandas
import pandas as pd
```

```python
#Johnson and Johnson
johnson = pd.read_csv('JNJ.csv')
johnson.head()
```

```python
pfizer = pd.read_csv('PFE.csv')
pfizer.head()
```

```python
roche = pd.read_csv('RHHBY.csv') #Roche Holding AG
roche.head()
```

```python
astrazeneca = pd.read_csv('AZN.csv')
astrazeneca.head()
```

```python
united = pd.read_csv('UNH.csv')
united.head()
```

```python
dfs = [johnson, pfizer, roche, astrazeneca, united]
```

```python
for df in dfs:
    df['MA50'] = df.Close.rolling(50).mean()
    df['MA200'] = df.Close.rolling(200).mean()
```

```python
johnson.head(200)
```

```python
for df in dfs:
    df['Previous day close price'] = df.Close.shift(1)
```

```python
johnson.head()
```

```python
for df in dfs:
    df['Change in Price'] = df['Close'] - df['Previous day close price']
```

```python
johnson.head()
```

```python
for df in dfs:
    df['Percent change in price'] = df.Close.pct_change()
```

```python
johnson.head()
```

```python
for df in dfs:
    df['Previous day volume'] = df.Volume.shift(1)
```

```python
johnson.head()
```

```python
for df in dfs:
    df['Change in volume'] = df['Volume'] - df['Previous day volume']
```

```python
johnson.head()
```

```python
for df in dfs:
    df['Percent change in volume'] = df.Volume.pct_change()
```

```python
johnson.head()
```

```python
johnson.to_csv('Johnson.csv')
pfizer.to_csv('Pfizer.csv')
roche.to_csv('Roche.csv')
astrazeneca.to_csv('AstraZeneca.csv')
united.to_csv('UnitedHealth.csv')
```
